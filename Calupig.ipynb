```python
import pandas as pd # Data Manipulation
import numpy as np  # Numerical Operation
import matplotlib.pyplot as plt  # Data Visualisation
from sklearn.linear_model import LinearRegression   #Data Modelling
```

```python
data = pd.read_csv('/content/Lesson_4_student_grades_dataset.csv')
```

```python
data.head()
```

```python
X = data [['Hours_of_Study']] #Independent Variable
Y = data [['Grades']] #Dependent Variable
```

```python
model = LinearRegression()
model.fit(X,Y)
```

```python
slope = model.coef_[0]

intercept = model.intercept_

print('Slope',slope)
print('Interceplt',intercept)
```

```python
plt.scatter(X,Y)
plt.plot(X,model.predict(X),color = 'pink')
plt.xlabel("Independent Variable(X)")
plt.ylabel("Dependent Variable(Y)")
plt.title("Scatter Points of Data")
```
